```python
import pandas as pd
import nltk
import pickle
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
import tensorflow as tf
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model

nltk.download('punkt')
nltk.download('stopwords')

stop_words = set(stopwords.words('english'))

vocab_size = 50000
embedding_dim = 100
max_length = 50

def preprocess_text(text):
    if type(text) == str:
        tokens = word_tokenize(text.lower())
        filtered_tokens = [word for word in tokens if word.isalnum() and word not in stop_words]
        return ' '.join(filtered_tokens)
    else:
        return str(text)


model_100epoch = load_model('Model/test.h5')
tokenizer_filename = "Model/tokenizertest.pkl"
with open(tokenizer_filename, 'rb') as tokenizer_file:
    loaded_tokenizer = pickle.load(tokenizer_file)
```

```python
predict_df = pd.read_csv("Data/predict.csv")

predict_df['tweet_processed'] = predict_df['tweet'].apply(preprocess_text)
sequences = loaded_tokenizer.texts_to_sequences(predict_df['tweet_processed'])
padded_sequences = pad_sequences(sequences, maxlen=max_length, truncating='post')

predict_df['100epoch_model'] = model_100epoch.predict(padded_sequences)

predict_df.to_csv("Data/result.csv", index=False)
print("result saved!")
```

### **TESTING WITH USER INPUT**

```python
def predict_sentiment(tweet,model):
    preprocessed_tweet = preprocess_text(tweet)
    tweet_sequence = loaded_tokenizer.texts_to_sequences([preprocessed_tweet])
    padded_sequence = pad_sequences(tweet_sequence, maxlen=max_length, padding='post', truncating='post')
    sentiment_prediction = model.predict(padded_sequence)[0][0]
    return sentiment_prediction

for i in range(0,3):
    user_tweet = input("Enter a tweet : ")
    print(user_tweet)
    sentiment_prediction = predict_sentiment(user_tweet,model_100epoch)
    print("Sentiment prediction for model  :", sentiment_prediction)
```
