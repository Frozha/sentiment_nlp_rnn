```python
import nltk
import pandas as pd
from nltk.corpus import words
from nltk.sentiment.vader import SentimentIntensityAnalyzer
import matplotlib.pyplot as plt
import numpy as np

nltk.download('words')
english_words = set(words.words())

def is_english_word(word):
    return word.lower() in english_words

def countWords(string):
  strList = string.split()
  length = len(strList)
  return length

def sentenceScore(string):
  strList = string.split()
  count = 0
  for each in strList:
    if is_english_word(each):
      count = count + 1
  return count
```

```python
newData = {'Sentence':[],'No_of_Words':[],'Valid_Words_Score':[],'percentageEnglishWord':[]}
newDf = pd.read_csv('Raw_Youtube_Search_Results.csv')
```

```python
for each in newDf['Comment']:
  wordLength = countWords(each)
  print(wordLength)
  score = sentenceScore(each)
  print(score)
  if wordLength!=0:
    percentage = score/wordLength *100
  else :
    percentage = 0
  newData['Sentence'].append(each)
  newData['No_of_Words'].append(wordLength)
  newData['Valid_Words_Score'].append(score)
  newData['percentageEnglishWord'].append(percentage)
```

```python
dataFrame = pd.DataFrame(newData)
```

```python
dataFrame #Selection criteria for the comments to be filtered are set. Those words that are 100% english
#can be selected for fitting into the model.
```

```python
def data_prep(dataFrame, percent):
    filteredComments = []
    for i in range(len(dataFrame)):
        if dataFrame['percentageEnglishWord'].iloc[i] <= percent:
            val = dataFrame['Sentence'].iloc[i]
            filteredComments.append(val)

    return filteredComments

```

```python
percentages = [10,20,30,40,50,60,70,90,100]
dataframes = [] #Will contain 10 entries each with the filtered comment level of the above percentage
for percent in percentages:
    filtered_data = data_prep(dataFrame, percent)
    df = pd.DataFrame({'FilteredComments': filtered_data})
    dataframes.append(df)
```

```python
for each in range(len(dataframes)):
  print(dataframes[each])
```

```python
# Function to perform sentiment analysis using VADER
def perform_sentiment_analysis(data):
    sia = SentimentIntensityAnalyzer()
    sentiment_scores = [sia.polarity_scores(comment) for comment in data['FilteredComments']]
    return sentiment_scores


percentages = [10, 20, 30, 40, 50, 60, 70, 90, 100]
dataframes = []

for percent in percentages:
    filtered_data = data_prep(dataFrame, percent)
    df = pd.DataFrame({'FilteredComments': filtered_data})

    sentiment_scores = perform_sentiment_analysis(df)
    df['SentimentScores'] = sentiment_scores

    dataframes.append(df)

# Creating bar graphs for each dataframe
for i, df in enumerate(dataframes):
    sentiment_scores = df['SentimentScores']

    # Extracting the compound sentiment scores
    compound_scores = [score['compound'] for score in sentiment_scores]

    plt.figure(figsize=(8, 6))
    plt.bar(range(len(compound_scores)), compound_scores)
    plt.xlabel('Comment Index')
    plt.ylabel('Sentiment Score')
    plt.title(f'Sentiment Scores for DataFrame {i+1}')
    plt.show()
```

```python



def plot_sentiment_bars(sentiment_scores):
    labels = ['Positive', 'Neutral', 'Negative']
    values = [0, 0, 0]

    for score in sentiment_scores:
        compound_score = score['compound']
        if compound_score > 0.05:
            values[0] += 1  # Positive
        elif compound_score < -0.05:
            values[2] += 1  # Negative
        else:
            values[1] += 1  # Neutral

    plt.bar(labels, values)
    plt.xlabel('Sentiment')
    plt.ylabel('Count')
    plt.title('Sentiment Distribution')
    plt.show()

percentages = [10, 20, 30, 40, 50, 60, 70, 90, 100]
dataframes = []

for percent in percentages:
    filtered_data = data_prep(dataFrame, percent)
    df = pd.DataFrame({'FilteredComments': filtered_data})

    sentiment_scores = perform_sentiment_analysis(df)
    df['SentimentScores'] = sentiment_scores

    dataframes.append(df)

# Creating sentiment bar graphs for each dataframe
for i, df in enumerate(dataframes):
    sentiment_scores = df['SentimentScores']
    plot_sentiment_bars(sentiment_scores)
```
